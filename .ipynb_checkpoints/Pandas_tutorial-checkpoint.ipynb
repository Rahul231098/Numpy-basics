```python
import pandas as pd
```

```python
info = {
    "Name" : ["Rahul", "Harshita", "Shradha"],
    "CGPA" : [9.5, 7.5, 8.5]
}

df = pd.DataFrame(info)

print(df)
```

```python
s = pd.Series([1,2,3,4,5])
print(s)
print (type(s))

#index
print(s[0])
print(s[2])
```

```python
import pandas as pd
s = pd.Series([23, 24, 25, 26], index=["Adam", "Marko", "Niklas", "Ronaldo"])
print(s)

print(s["Niklas"]) # label access
 print(s[2]) # index access
```

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])

print(s1+s2)

s1[0] = 100
changed_s1 = s1.drop(0)

print(s1)
print(changed_s1)  # Hence, Series is immutable 
```

```python
#DataFrame

info = {
    "Name" : ["Adam", "Eve", "Bob"],
    "Age" : [23,24,25],
    "GPA" : [9.5,8.6,7.2]
}

df = pd.DataFrame(info)

print(df, type(df))

print(df.index)
print(df.columns)
```

```python
df
```

```python
df = pd.DataFrame([["Adam",23], ["Bob",23], ["Eve",23]], columns = ["Name", "Age"])

print(df)
```

```python
import numpy as np

np_array = np.array([[1,2,3],[4,5,6]])

df = pd.DataFrame(np_array, columns=["A","B","C"])

print(df)
```
